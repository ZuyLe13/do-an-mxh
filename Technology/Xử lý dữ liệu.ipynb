```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
import numpy as np
```

```python
def load_json_to_dataframe(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line_num, line in enumerate(f, start=1):
            try:
                data.append(json.loads(line.strip()))
            except json.JSONDecodeError as e:
                print(f"Skipping malformed line {line_num} in {file_path}: {e}")
    return pd.DataFrame(data)

# Sử dụng hàm cho các file
user = load_json_to_dataframe('user.json')
video = load_json_to_dataframe('video.json')
course = load_json_to_dataframe('course.json')
problem = load_json_to_dataframe('problem.json')
user_video = load_json_to_dataframe('user-video.json')
user_problem = load_json_to_dataframe('user-problem.json')

print("Data loaded successfully!")
```

```python
# Loại bỏ tiền tố 'C_' trong cột id
course['id'] = course['id'].str.replace('C_', '', regex=False)
course = course.drop(columns=['field', 'prerequisites', 'about'])
course.head()
```

```python
# Tách video_id và ex_id từ cột resource
course['video'] = course['resource'].apply(lambda x: [item['resource_id'] for item in x if item['resource_id'].startswith("V_")])
course['ex'] = course['resource'].apply(lambda x: [item['resource_id'] for item in x if item['resource_id'].startswith("Ex_")])

# Loại bỏ cột resource
course = course.drop(columns=['resource'])

# Hiển thị kết quả
course.head(10)
```

```python
course['total_videos'] = course['video'].apply(len)
course['total_exercises'] = course['ex'].apply(len)
course.head(10)
```

```python
course.to_csv('course_cleaned.csv', index=False, encoding='utf-8')
```
